```python
X1 = df_copy.drop("FWI",axis=1)
Y2 = df_copy['FWI']
```

```python
from sklearn.model_selection import train_test_split
x_train2, x_test2, y_train2, y_test2 = train_test_split(X1,Y2,test_size=0.25, random_state=42)
```

```python
x_train2.corr()
```

```python
plt.figure(figsize=(12,10))
corr = x_train2.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
def correlation(dt, threshold):
    col_corr = set()
    corr_matrix = dt.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname= corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features=correlation(x_train2,0.85)
```

```python
x_train2.drop(corr_features,axis=1,inplace=True)
x_test2.drop(corr_features,axis=1,inplace=True)
x_train2.shape,x_test2.shape
```

```python
from sklearn.preprocessing import StandardScaler 
scalar = StandardScaler() 
x_train_scaled = scalar.fit_transform(x_train2) 
x_test_scaled = scalar.transform(x_test2)
```

##Linear Regression

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
reg = LinearRegression() 
reg.fit(x_train_scaled,y_train2) 
y_pred_new = reg.predict(x_test_scaled)
mae = mean_absolute_error(y_test2,y_pred_new)
score = r2_score(y_test2,y_pred_new)
print("mean_absolute_error:",mae)
print("Accuracy score:",score)
plt.scatter(y_test2,y_pred_new)
plt.show()
```

##Lasso Regresion

```python
from sklearn.linear_model import Lasso 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
reg_lasso = Lasso() 
reg_lasso.fit(x_train_scaled,y_train2) 
y_pred_new = reg_lasso.predict(x_test_scaled)
mae = mean_absolute_error(y_test2,y_pred_new)
score = r2_score(y_test2,y_pred_new)
print("mean_absolute_error:",mae)
print("Accuracy score:",score)
plt.scatter(y_test2,y_pred_new)
plt.show()
```

##Cross Validation Lasso

```python
from sklearn.linear_model import LassoCV 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
reg_lassoCV = LassoCV() 
reg_lassoCV.fit(x_train_scaled,y_train2) 
y_pred_new = reg_lassoCV.predict(x_test_scaled)
mae = mean_absolute_error(y_test2,y_pred_new)
score = r2_score(y_test2,y_pred_new)
print("mean_absolute_error:",mae)
print("Accuracy score:",score)
plt.scatter(y_test2,y_pred_new)
plt.show()
```

##Ridge Regression

```python
from sklearn.linear_model import Ridge 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
reg_Ridge = Ridge() 
reg_Ridge.fit(x_train_scaled,y_train2) 
y_pred_new = reg_Ridge.predict(x_test_scaled)
mae = mean_absolute_error(y_test2,y_pred_new)
score = r2_score(y_test2,y_pred_new)
print("mean_absolute_error:",mae)
print("Accuracy score:",score)
plt.scatter(y_test2,y_pred_new)
plt.show()
```
